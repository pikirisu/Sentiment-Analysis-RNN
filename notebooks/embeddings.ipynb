```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
#sentences
sent = [
    'the glass of milk',
    'the glass of blood',
    'the cup of sause',
    'I am the demon',
    'I am the winner',
    'I am the butcher of dreams',
    'understand the meaning of these words',
    'I was always ahead',
]
```

```python
sent    
```

```python
# Vocab_size
voc_size = 10000
```

```python
#One Hot representation
one_hot_repr = [one_hot(words,voc_size) for words in sent]
one_hot_repr
# Indexes of words described above
```

```python
# Word Embeddings
from tensorflow.keras.layers import Embedding,Input
# from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
# Pad sequence to pad words to make them of equal size --> pre and post in padding for 0s in beg /end
sen_length = 10
embedded_docs = pad_sequences(one_hot_repr,padding='pre',maxlen=sen_length)
embedded_docs
```

```python
embedded_docs.dtype
```

```python
# feature representation ( can keep 300 from google one)
dim =10
```

```python
model = Sequential()
model.add(Input(shape=(sen_length,)))
model.add(Embedding(input_dim=voc_size,output_dim=dim))
model.compile(optimizer='adam',loss='mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
